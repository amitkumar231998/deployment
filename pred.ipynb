```python
from sklearn.preprocessing import scale
df_scale=scale(data)
```

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
import numpy as np

pca=PCA(n_components=(35)) #Principal component on 35 columns
pca_values=pca.fit_transform(df_scale)
var=pca.explained_variance_ratio_
var1=np.cumsum(np.round(var,decimals=4)*100)
print("Variance Ratio: ",var1)
```

```python
pca_values
type(pca_values)
pca_data = pd.DataFrame(pca_values)

#Naming the PCA columns
name=[]
for i in range(1,36):
    a="PC"+str(i)
    name.append(a)
pca_data.columns =name 

#Considering only 26 PCA as they it summraise the 91% of the data, so dropping 9 PCA
final = pd.concat([pca_data.iloc[:, 0:26],data.Renewal], axis = 1)
```

```python
x=data.iloc[:,0:25]
y=data["Renewal"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25)

from sklearn.tree import DecisionTreeClassifier as DT
model_decision = DT(criterion = 'entropy')
model_decision.fit(x_train,y_train)
```

```python
pred=model_decision.predict(x_test)
```
